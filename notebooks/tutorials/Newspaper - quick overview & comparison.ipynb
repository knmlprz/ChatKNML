## Porównanie zwykłego scrapowania vs scrapowania z użyciem newspaper

```python
# Przypisanie adresu URL do zmiennej

url = 'https://w.prz.edu.pl/studenci_menu/legitymacje'
```

```python
# Import modułu requests oraz obiektu BeautifulSoup

from bs4 import BeautifulSoup as Soup
import requests

# Pobranie zawartości strony

page = requests.get(url)

# Przypisanie zawartości strony w formacie string do zmiennej 

full_text = Soup(page.content, "html.parser").text
```

### Tak prezentują się zaśmiecone dane, z których chcemy wydobyć wartościowe treści

```python
print(full_text)
```

### Z pomocą przychodzi nam moduł newspaper

```python
print()
```

```python
# Import obiektu Article z modułu newspaper

from newspaper import Article

# Utworzenie obiektu Article

article = Article(url, language='pl')

# Wywołanie metody download, która pobiera źródło strony

article.download()

# Wywołanie metody parse, która parsuje dane do odpowiedniego formatu

article.parse()

# Przypisanie zawartości strony w formacie string do miennej 

clean_text = article.text
```

### Tak prezentują się dane, które zawierają wyłącznie wartościowe treści

```python
print(clean_text)
```

### Jak bardzo skrócił się tekst?

```python
print(f'''Długość pełnego tekstu strony: {len(full_text)}
Długość istotnego tekstu strony: {len(clean_text)}

Różnica: {len(full_text) - len(clean_text)}''')
```

### Dodatkowe funkcje modułu newspaper

```python
# Wywołanie metody nlp, która tworzy dodatkowe wartości obiektu

article.nlp()
```
